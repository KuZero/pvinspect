````python
from pvinspect.data import read_module_images, EL_IMAGE
````

[```read_module_images```](https://ma0ho.github.io/pvinspect/data/index.html#pvinspect.data.read_module_images) is used to read data from the disk. You need to specify the data path, imaging modality and if the images originate from the same camera such that intrinsic parameters are the same for all images. In addition, you might also specify the number of cells in a row and in a column. For more information and additional arguments, please refer to the [documentation](https://ma0ho.github.io/pvinspect/data/index.html#pvinspect.data.read_module_images).

````python
imgs = read_module_images('../pvinspect/data/datasets/20191219_poly10x6', EL_IMAGE, same_camera=True, cols=10, rows=6)
````

We can now proceed with the resulting [```ModuleImageSequence```](https://ma0ho.github.io/pvinspect/data/image.html#pvinspect.data.image.ModuleImageSequence)..:
